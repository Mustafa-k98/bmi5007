```python
phys_df = syn_df.loc[:, ["DESYNPUF_ID", "AT_PHYSN_NPI", "OP_PHYSN_NPI", "OT_PHYSN_NPI"]]
```

```python
phys_df.groupby("DESYNPUF_ID").agg({"AT_PHYSN_NPI":'count', "OP_PHYSN_NPI":'count', "OT_PHYSN_NPI":'count'})
```

```python
def get_unique(group):
    physicians_col = [group["AT_PHYSN_NPI"],group["OP_PHYSN_NPI"],group["OT_PHYSN_NPI"]]
    return len(pd.concat(physicians_col).dropna(inplace=False).unique())
    
```

```python
syn_df.groupby("DESYNPUF_ID").apply(get_unique)
```
