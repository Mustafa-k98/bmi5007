```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%matplotlib inline
```

```python
syn_df = pd.read_csv("data/synpuf.csv")
```

```python
syn_df
```

```python
syn_df.columns
```

```python
def get_unique(group):
    icd9_cols=[group['ICD9_DGNS_CD_'+ str(k)] for k in range(1,11)]
    return list(pd.concat(icd9_cols, axis=0).dropna(inplace=False).unique().astype(str))
```

```python
icd9_df = syn_df.groupby("DESYNPUF_ID").apply(get_unique)
```

```python
icd9_df.explode()
```

```python
g=pd.get_dummies(icd9_df.explode()).groupby("DESYNPUF_ID").sum()
```

```python
sns.heatmap(g)
```

```python
import collections
```

```python
def get_count(group):
    icd9_cols=[group['ICD9_DGNS_CD_'+ str(k)] for k in range(1,11)]
    return collections.Counter(pd.concat(icd9_cols, axis=0).dropna(inplace=False).astype(str))
```

```python
icd9_count_df = syn_df.groupby("DESYNPUF_ID").apply(get_count)
```

```python
g2 = icd9_count_df.apply(pd.Series).fillna(0)
```

```python
sns.heatmap(g2)
```
